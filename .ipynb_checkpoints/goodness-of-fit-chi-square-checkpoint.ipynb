# Chi-Square Goodness of Fit Test

Hypothesis test to check if the **real-world data** matches **what we expected**. This is often called **'Good of Fit'** or the **'Test of Independence'**.

**The Core Concept: Observed vs Expected**
The test compares two things.
1) **Observed ($O$)**: What actually happened in your experiment/ survey.
2) **Expected ($E$)**: What mathematically should have happened if you hypothesis was true (e.g., if a coin was perfectly fair, you expect 50 heads in 100 flips)

**The Formula**

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

**The Intuition behind the Math:**
- **$(O - E)$**: Difference between reality and expectation
- **$(O - E)^2$**: Square it so that negatice differences don't cancel out positive ones. This emphasizes big mistakes.
- **Divide by $E$**: Divide by the expected value to scale it. (Missing a count by 10 matters a lot if you only expected 5, but it matters very little if you expected 100,000)

**How the test works:**
- **Null Hypothesis ($H0$)**: "Nothing special is happening. The data fits the pattern perfectly, and any difference is just a random luck
- **If $X^2$ is small**: The Observed is close to the Expected. You trust the Null Hypothesis
- **If $X^2$ is huge**: The Observed is very different from the Expected. You reject the Null Hypothesis.

### How big is too big

#### **Critical Value**:
- The **"Line in the Sand"**
- This is a pre-determined cutoff point. Before you even start your experiment, decide a **"Significance Level"** (usually 0.05 or 5%)
- **How it is Calculated?**: By looking up in a standard statistical table based on **Significance Level** (alpha), and **Degrees of Freedom (df)** (number of Categories -1)

#### **P-Value**:
- The **"The factor of Surprise"**
- This tells you the probability of getting you specific result just by luck.
- **How it is calculated?**: It is the area under the curve to the right of calculated Chi-square score. (Calculated by software/calculators, as the math involves integrals).
- **Intuition:**
  - High P-value (e.g., 0.80): "Not surprising at all. This happens all the time by chance." Keep Null Hypothesis.
  - Low P-value (e.g., 0.01): "Extremely surprising! Ideally, this rarely happens by chance. Something real must be causing this". Reject Null Hypothesis.

```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dice data.xlsx", sheet_name="Sheet1")
df
```

### Run the test

```python
deg_of_freedom = len(df)-1
alpha = 0.05 # The significance level (5%)

chi2_score, p_value = stats.chisquare(f_obs=df["Observed"], f_exp=df["Expected"])
critical_value = stats.chi2.ppf(1 - alpha, deg_of_freedom)

print("\n--------- RESULTS ----------\n")
print(f"Chi-Square Score: {chi2_score:.4f}")
print(f"P-Value: {p_value: .4f}")
print(f"Critical Value: {critical_value: .4f}")
```

### Interpretation of Result

```python
print("\n--------- CONCLUSION ----------\n")

if p_value < alpha:
    print("Result: Statistically Significant (Reject Null Hypothesis)")
    print("Interpretation: The dice is likely rigged")
else:
    print("Result: Not Significant (Fail to reject Null Hypothesis)")
    print("Interpretation: The dice is fair (Variations are just luck)")
```

### Visualize the Results

#### Expected vs Observed

```python
df_melted = df.melt(
    id_vars = "Face", 
    value_vars = ["Observed", "Expected"], 
    var_name = "Type", 
    value_name = "Count"
)

df_melted
```

```python
ax = sns.barplot(
    data = df_melted,
    x = "Face",
    y = "Count",
    hue = "Type",
    palette = "muted"
)

ax.set_title("Observed vs Expected Results")

for container in ax.containers:
    ax.bar_label(container)
```

#### Chi-square Distribution Curve

```python
x = np.linspace(0, 25, 1000) # range to see the tail
y = stats.chi2.pdf(x, deg_of_freedom)

# Visualization
ax2 = sns.lineplot(
    x = x,
    y = y,
    color = "black",
    label = "Chi-Square PDF"
)

# Show the Major grid
ax2.grid(which = "major", linestyle = "--", linewidth=0.5, alpha=0.7)

# Highlight the Critical Region (The area where we reject H0)
# This fills the area starting from the Critical Value to the right
ax2.fill_between(
    x, y, 
    where=(x > critical_value), 
    color="orange", 
    alpha=0.3, 
    label=f"Critical Region (alpha={alpha})"
)

# Highlight "Our Result" (The P-Value area)
# This fills the area starting from your calculated Score to the right
ax2.fill_between(
    x, y, 
    where=(x > chi2_score), 
    color="blue", 
    alpha=0.3, 
    label=f"P-Value Region (p={p_value:.3f})"
)

# Vertical Line for Critical Value (The Threshold)
ax2.axvline(
    x=critical_value, 
    color="orange", 
    linestyle="--", 
    linewidth=2, 
    label=f"Critical Value: {critical_value:.2f}"
)

# Vertical Line for Our Result (Observed Statistic)
ax2.axvline(
    x=chi2_score, 
    color="blue", 
    linestyle="-", 
    linewidth=2, 
    label=f"Our Result: {chi2_score:.2f}"
)

# Formatting
ax2.set_title(f"Chi-Square Distribution (df={deg_of_freedom})", fontsize=14)
ax2.set_xlabel("Chi-Square Statistic")
ax2.set_ylabel("Probability Density")
ax2.legend()

plt.show()
```
